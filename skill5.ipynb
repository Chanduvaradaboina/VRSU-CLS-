```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python


# Read the image
image = cv2.imread('/content/Screenshot 2026-01-02 140849.png')   # replace with your image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur
blur = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply Canny Edge Detection
edges = cv2.Canny(blur, threshold1=50, threshold2=150)

# Display results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread('/content/Screenshot 2026-01-02 140849.png')   # replace with your image path
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create SIFT object
sift = cv2.SIFT_create()

# Detect keypoints and descriptors
keypoints, descriptors = sift.detectAndCompute(gray, None)

# Draw keypoints
sift_image = cv2.drawKeypoints(
    image,
    keypoints,
    None,
    flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS
)

# Convert to RGB for display
sift_image_rgb = cv2.cvtColor(sift_image, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(6,6))
plt.imshow(sift_image_rgb)
plt.title(f"SIFT Keypoints Detected: {len(keypoints)}")
plt.axis("off")
plt.show()
```
